# Deep Learning for Tick-Level Price Prediction
Training the `LSTMPricePredictor` using PyTorch on historical Level-2 micro-structure data.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Import model from our repository
import sys; sys.path.append('..')
from ml_models.lstm_price_predictor import LSTMPricePredictor
```

## Simulated DataLoader & Training Loop
Since this is an example notebook, we simulate the dataloader generating 10-feature sequences.

```python
model = LSTMPricePredictor(input_size=10, hidden_layer_size=64, output_size=1)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 50
losses = []

for epoch in range(epochs):
    # Simulated batch: [Batch Size, Sequence Length, Features]
    inputs = torch.randn(32, 5, 10)
    targets = torch.randn(32, 1)
    
    optimizer.zero_grad()
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    loss.backward()
    optimizer.step()
    
    losses.append(loss.item())

plt.plot(losses)
plt.title('LSTM Training Loss Convergence')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.show()
```
