# Order Book Visualization & Order Flow Imbalance (OFI)
This notebook explores Level-2 market data micro-structure. We visualize the Limit Order Book (LOB) depth and calculate the Order Flow Imbalance to predict short-term price movements.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## Loading Snapshots
We simulate loading high-frequency parquet snapshots.

```python
# Simulated LOB snapshot data
bids = pd.DataFrame({'price': [50000 - i*0.5 for i in range(50)], 'volume': np.random.exponential(1.5, 50)})
asks = pd.DataFrame({'price': [50000 + i*0.5 for i in range(50)], 'volume': np.random.exponential(1.5, 50)})

plt.fill_between(bids['price'], bids['volume'], step='post', alpha=0.5, color='green', label='Bids')
plt.fill_between(asks['price'], asks['volume'], step='post', alpha=0.5, color='red', label='Asks')
plt.title('Level-2 Limit Order Book Depth')
plt.xlabel('Price ($)')
plt.ylabel('Volume')
plt.legend()
plt.show()
```

## Order Flow Imbalance (OFI)
OFI tracks the net change in resting order volumes at the best bid and ask over time. A positive OFI suggests buying pressure.

```python
def calculate_ofi(bid_vol_t, ask_vol_t, bid_vol_prev, ask_vol_prev):
    delta_w = bid_vol_t if bid_vol_t >= bid_vol_prev else 0
    delta_v = ask_vol_t if ask_vol_t >= ask_vol_prev else 0
    return delta_w - delta_v

# Sample OFI over 100 ticks
ofi_series = np.random.randn(100).cumsum()
plt.plot(ofi_series, color='purple')
plt.title('Order Flow Imbalance (OFI) over Time')
plt.show()
```
