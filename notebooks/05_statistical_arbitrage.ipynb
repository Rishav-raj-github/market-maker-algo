# Cointegration Analysis for Statistical Arbitrage
Validating pairs trading signals using the Augmented Dickey-Fuller (ADF) test.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
```

## Generating Synthetically Cointegrated Price Series
We create a random walk for Asset A, and make Asset B follow it with some mean-reverting noise.

```python
np.random.seed(42)
A = np.cumsum(np.random.randn(1000)) + 100
noise = np.random.randn(1000) * 0.5
B = A + noise + 5.0 # B tracks A with a premium

plt.plot(A, label='Asset A')
plt.plot(B, label='Asset B')
plt.legend()
plt.title('Cointegrated Asset Prices')
plt.show()
```

## ADF Test & Z-Score Calculation
If the p-value is < 0.05, the spread is stationary, allowing us to trade the Z-score.

```python
spread = B - A
adf_result = adfuller(spread)
print(f"ADF Statistic: {adf_result[0]:.4f}")
print(f"p-value: {adf_result[1]:.4e}")

z_score = (spread - np.mean(spread)) / np.std(spread)
plt.plot(z_score, color='purple')
plt.axhline(2, color='red', linestyle='--')
plt.axhline(-2, color='green', linestyle='--')
plt.title('Spread Z-Score')
plt.show()
```
