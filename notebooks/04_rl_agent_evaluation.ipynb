# Evaluating the Reinforcement Learning Market Maker
We evaluate the `stable-baselines3` PPO agent on unseen simulated market data.

```python
import numpy as np
import matplotlib.pyplot as plt
from stable_baselines3 import PPO
from strategy.rl_market_maker import MarketMakingEnv
import sys; sys.path.append('..')
```

```python
env = MarketMakingEnv()
# Assuming the model was saved as 'rl_market_maker_v1'
# model = PPO.load('../rl_market_maker_v1')

pnl_history = [0]
inventory_history = []

# Simulated evaluation loop
for step in range(500):
    action = env.action_space.sample() # Random action for baseline comparison
    obs, reward, done, _ = env.step(action)
    pnl_history.append(pnl_history[-1] + reward)
    inventory_history.append(obs[0])

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
ax1.plot(pnl_history, color='green')
ax1.set_title('Cumulative PnL over Episode')
ax2.plot(inventory_history, color='orange')
ax2.set_title('Inventory State Tracking')
plt.tight_layout()
plt.show()
```
