## 1. The Reservation Price
The reservation price $r(s, t)$ is the ideal price at which the market maker wants to center their quotes, adjusting for inventory risk.

$$ r(s, t) = s - q \cdot \gamma \cdot \sigma^2 \cdot (T - t) $$

Where:
- $s$ = current mid-price
- $q$ = current inventory
- $\gamma$ = risk aversion parameter
- $\sigma$ = volatility of the asset
- $T - t$ = time left to terminal state

```python
import numpy as np
import matplotlib.pyplot as plt

def reservation_price(s, q, gamma, sigma, time_left):
    return s - q * gamma * (sigma ** 2) * time_left

inventories = np.arange(-10, 11, 1)
r_prices = [reservation_price(100, q, 0.1, 1.5, 1.0) for q in inventories]

plt.plot(inventories, r_prices, marker='o')
plt.axhline(100, color='r', linestyle='--', label='Mid Price')
plt.title('Reservation Price vs Inventory')
plt.xlabel('Inventory (q)')
plt.ylabel('Reservation Price (r)')
plt.legend()
plt.show()
```

## 2. Optimal Spread
The optimal spread $\delta$ balances earning the spread against adverse selection risk:

$$ \delta = \gamma \sigma^2 (T - t) + \frac{2}{\gamma} \ln\left(1 + \frac{\gamma}{k}\right) $$

```python
def optimal_spread(gamma, sigma, time_left, k):
    return gamma * (sigma ** 2) * time_left + (2 / gamma) * np.log(1 + gamma / k)

print(f"Optimal spread at t=0: {optimal_spread(0.1, 1.5, 1.0, 1.5):.4f}")
```
